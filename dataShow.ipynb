```python
def mostraImagem(link, imagem):
  image_url = link
  image_path = imagem
  urllib.request.urlretrieve(image_url, image_path)
  image = plt.imread(image_path)
  plt.figure(figsize=(10, 10))
  plt.imshow(image)
  plt.axis('off')
  plt.show()
```

```python
df = pd.read_csv('dados.csv', delimiter=',')
print(df.columns)
```

#Dados:

##Total de cada palavra





```python
keyword_counts = df['keyWord'].value_counts()
print(keyword_counts)
print("total = ", keyword_counts.sum())
```

```python
plt.figure(figsize=(8, 6))
plt.pie(keyword_counts, labels=keyword_counts.index, autopct='%1.1f%%')
plt.title('Keyword Distribution')
plt.axis('equal')
plt.show()
```

##Data mais antiga de cada palavra

```python
oldest_dates = df.groupby('keyWord')['dateTime'].min()

# Print the oldest date for each keyword
for keyword, date in oldest_dates.items():
    print(f"Oldest date for '{keyword}': {date}")
```
